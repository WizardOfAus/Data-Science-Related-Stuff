##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def Std_Dev(List_Inp):
    List_mean=(sum(List_Inp))/(len(List_Inp))
    Variance=sum([(x-List_mean)**2 for x in List_Inp])/len(List_Inp)
    Standard_Deviation=Variance**(1/2)
    return Variance,Standard_Deviation
print "The Variance and Standard Deviation of the Participation Rate is %s and %s respectively." %(Std_Dev(Overall_Dictionary['Rate']))
print "The Variance and Standard Deviation of the Math Scores are %s and %s respectively." %Std_Dev(Overall_Dictionary['Math'])
print "The Variance and Standard Deviation of the Verbal Scores are %s and %s respectively." %Std_Dev(Overall_Dictionary['Verb'])
```
